```python
import numpy as np
import pandas as pd
pd.set_option("display.max.columns", 100)
%matplotlib inline
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings("ignore")
```

```python
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"
```

```python
data = pd.read_csv(DATA_URL + "adult.data.csv")
data.head()
```

```python
#Q1
data["sex"].value_counts()
```

```python
#Q2
data[data["sex"]=="Female"]["age"].mean()
```

```python
#Q3
float((data["native-country"] == "Germany").sum()) / data.shape[0]
```

```python
#Q4
moreavg = data[data["salary"]==">50K"]["age"].mean()
print(f">50k mean age: {moreavg}")
lessavg = data[data["salary"]=="<=50K"]["age"].mean()
print(f"<=50k mean age: {lessavg}")
#Q5
moredev = data[data["salary"]==">50K"]["age"].std()
print(f">50k mean age: {moredev}")
lessdev = data[data["salary"]=="<=50K"]["age"].std()
print(f"<=50k mean age: {lessdev}")
```

```python
#Q6
data[data["salary"]==">50K"]["education"].unique()
```

```python
#Q7
columnstoshow = ["race", "sex"]
for (race, sex), subdf in data.groupby(columnstoshow):
  print("Race: {0}, sex: {1}".format(race, sex))
  print(subdf["age"].describe())
```

```python
#Q8
data[data["marital-status"].str.startswith("Married")][data["sex"]=="Male"]["salary"].value_counts(normalize=True)
```

```python
data[~data["marital-status"].str.startswith("Married")][data["sex"]=="Male"]["salary"].value_counts(normalize=True)
```

```python
maxh = data["hours-per-week"].max()
print("max hours per week: " + str(maxh))
print("number of people: " + str((data["hours-per-week"] == maxh).sum()))
numrich = data[(data["hours-per-week"] == maxh) & (data["salary"] == ">50K")].shape[0]
print(numrich/(data["hours-per-week"] == maxh).sum())
```
